```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

import warnings
warnings.filterwarnings("ignore")

# File to Load 
winter_load = Path("Citibike data files/JC-201801-citibike-tripdata.csv")
spring_load = Path("Citibike data files/JC-201804-citibike-tripdata.csv")
summer_load = Path("Citibike data files/JC-201807-citibike-tripdata.csv")
fall_load = Path("Citibike data files/JC-201810-citibike-tripdata.csv")

# Read data files and store into Pandas DataFrames
winter_data = pd.read_csv(winter_load)
spring_data = pd.read_csv(spring_load)
summer_data = pd.read_csv(summer_load)
fall_data = pd.read_csv(fall_load)

# # Combine the data into a single dataset.  
# school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
# school_data_complete.head()
```

```python
#merge the dataframes
merged1 = winter_data.merge(spring_data, how='outer')
merged1.head(3)
```

```python
merged2 = merged1.merge(summer_data, how='outer')
merged2.head(2)
```

```python
merged_complete = merged2.merge(fall_data, how='outer')
merged_complete.head(2)
```

```python
merged_complete.info()
```

```python
# change gender to male, female, or unknown
merged_complete['gender'] = merged_complete['gender'].replace([0,1,2],['unknown','male', 'female'])
```
